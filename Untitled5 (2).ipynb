## here i want to see the total number of matches each team played in order to see who was playing in the final
### the two teams with the most played games are the teams in the final

```python
team1_counts = df['team1'].value_counts()
team2_counts = df['team2'].value_counts()
total_matches = team1_counts.add(team2_counts, fill_value=0).astype(int)
Teams = pd.concat([df['team1'], df['team2']])
total_matches_df = total_matches.reset_index()
total_matches_df.columns = ['Teams', 'Matches']
plt.figure(figsize=(26, 7))
plt.bar(total_matches_df['Teams'], total_matches_df['Matches'], color='blue')
plt.title('Total Goals Scored by Each Team in Euro 2024')
plt.xlabel('Teams')
plt.ylabel('Matches')
plt.show()
```

## This graph indicates the total goals scored by each team in the tournument

```python
team1_goals = df.groupby('team1')['team1_goals'].sum()
team2_goals = df.groupby('team2')['team2_goals'].sum()
total_goals = team1_goals.add(team2_goals, fill_value=0).astype(int)
total_goals_df = total_goals.reset_index()
total_goals_df.columns = ['Team', 'Goals_Scored']
teams = pd.concat([df['team1'], df['team2']])
plt.figure(figsize=(26, 7))
plt.bar(total_goals_df['Team'], total_goals_df['Goals_Scored'], color='blue')
plt.title('Total Goals Scored by Each Team in Euro 2024')
plt.xlabel('Team')
plt.ylabel('Goals Scored')
plt.show()
```

## this graph comapres between every team's total shot to scored goals.

```python
team1_goals = df.groupby('team1')['team1_goals'].sum()
team2_goals = df.groupby('team2')['team2_goals'].sum()
total_goals = team1_goals.add(team2_goals, fill_value=0).astype(int)
team1_shots = df.groupby('team1')['team1_total_shots'].sum()
team2_shots = df.groupby('team2')['team2_total_shots'].sum()
total_shots = team1_shots.add(team2_shots, fill_value=0).astype(int)
total_goals_df = total_goals.reset_index()
total_goals_df.columns = ['Team', 'Goals_Scored']
total_shots_df = total_shots.reset_index()
total_shots_df.columns = ['Team', 'Shots_Taken']
# Merge the dataframes on 'Team'
comparison_df = pd.merge(total_goals_df, total_shots_df, on='Team')
# Plot the data using matplotlib
plt.figure(figsize=(28, 10))
# Plot goals scored
plt.bar(comparison_df['Team'], comparison_df['Goals_Scored'], color='blue', alpha=0.6, label='Goals Scored')
# Plot shots taken
plt.bar(comparison_df['Team'], comparison_df['Shots_Taken'], color='grey', alpha=0.6, label='total shots', bottom=comparison_df['Goals_Scored'])
plt.title('Total Goals Scored vs. Total Shots Taken by Each Team in Euro 2024')
plt.xlabel('Team')
plt.ylabel('Total')
plt.legend()
plt.show()
```

## total shots on target compared to  total goals for each team

```python
team1_goals = df.groupby('team1')['team1_goals'].sum()
team2_goals = df.groupby('team2')['team2_goals'].sum()
total_goals = team1_goals.add(team2_goals, fill_value=0).astype(int)
team1_total_shots_on_target = df.groupby('team1')['team1_shots_on_target'].sum()
team2_total_shots_on_target = df.groupby('team2')['team2_shots_on_target'].sum()
total_shots_on_target = team1_total_shots_on_target.add(team2_total_shots_on_target, fill_value=0).astype(int)
total_goals_df = total_goals.reset_index()
total_goals_df.columns = ['Team', 'Goals_Scored']
total_shots_on_target_df = total_shots_on_target.reset_index()
total_shots_on_target_df.columns = ['Team', 'total_shots_on_target']
# Merge the dataframes on 'Team'
comparison_df = pd.merge(total_goals_df, total_shots_on_target_df, on='Team')
# Plot the data using matplotlib
plt.figure(figsize=(28, 10))
# Plot goals scored
plt.bar(comparison_df['Team'], comparison_df['Goals_Scored'], color='blue', alpha=0.6, label='Goals Scored')
# Plot shots taken
plt.bar(comparison_df['Team'], comparison_df['total_shots_on_target'], color='grey', alpha=0.6, label='total_shots_on_target', bottom=comparison_df['Goals_Scored'])
plt.title('Total Goals Scored vs. Total Shots on Target Taken by Each Team in Euro 2024')
plt.xlabel('Team')
plt.ylabel('Total')
plt.legend()
plt.show() 
```

## in this graph i wanted to see the percentage of total goals compared to the total shots taken in the tornument

```python
team1_goals = df.groupby('team1')['team1_goals'].sum()
team2_goals = df.groupby('team2')['team2_goals'].sum()
total_goals = team1_goals.add(team2_goals, fill_value=0).astype(int).sum()

# Calculate total shots for each team
team1_total_shots = df.groupby('team1')['team1_total_shots'].sum()
team2_total_shots = df.groupby('team2')['team2_total_shots'].sum()
total_shots = team1_total_shots.add(team2_total_shots, fill_value=0).astype(int).sum()

# Create a pie chart
labels = ['Total Goals Scored', 'Total Shots Taken']
sizes = [total_goals, total_shots]
colors = ['blue', 'orange']
explode = (0.1, 0)  # explode the 1st slice (Total Goals Scored

plt.figure(figsize=(6, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=180)
plt.title('Total Goals Scored vs. Total Shots Taken in Euro 2024')
plt.legend();
plt.show()
```

 ## here i wanted to see the total goals and and avergae goal per match to  decide if its high or normal or low

```python
team1_goals = df['team1_goals'].sum()
team2_goals = df['team2_goals'].sum()
total_goals = team1_goals + team2_goals

# Calculate the total number of matches played
total_matches = df.shape[0]

# Calculate the average number of goals scored per match
average_goals_per_match = total_goals / total_matches

# Create a DataFrame for visualization
visualization_df = pd.DataFrame({
    'Metric': ['Total Goals', 'Average Goals per Match'],
    'Value': [total_goals, average_goals_per_match]
})

# Plot the data using matplotlib
plt.figure(figsize=(10, 6))
plt.plot(visualization_df['Metric'], visualization_df['Value'], marker='o', linestyle='-', color='blue')
plt.title('Total Goals and Average Goals per Match in Euro 2024')
plt.xlabel('Metric')
plt.ylabel('Value')
plt.ylim(0, max(visualization_df['Value']) + 10)
plt.yticks(range(0, int(max(visualization_df['Value'])) + 5,5))
plt.show()
```
